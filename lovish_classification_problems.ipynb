```python
import pandas as pd

def extract_relevant_data(input_file, output_file):
    """
    Extract relevant data from the CSV file where negativereason is not null,
    keeping only text, negativereason, and airline_sentiment columns.
    """
    # Read the CSV file
    df = pd.read_csv(input_file)

    # Filter for rows where negativereason is not null
    filtered_df = df[df['negativereason'].notnull()].copy()

    # Select only the required columns
    result_df = filtered_df[['text', 'negativereason', 'airline_sentiment']]

    # Save results
    result_df.to_csv(output_file, index=False)
    print(f"Extraction complete. Results saved to {output_file}")

    return result_df

if __name__ == "__main__":
    input_csv = "/content/Tweets.csv"  #  input file path
    output_csv = "filtered_tweets.csv"  # Output file path

    # Process the data
    extracted_data = extract_relevant_data(input_csv, output_csv)

    # Print summary
    print("\nExtracted Data Summary:")
    print(f"Total rows extracted: {len(extracted_data)}")
```
